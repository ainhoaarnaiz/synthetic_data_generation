```python
import cv2
import json
import matplotlib.pyplot as plt
import os

script_dir = os.getcwd()
img_path = os.path.join(script_dir, "..", "render", "0001.png")
json_path = os.path.join(script_dir, "..", "render", "0001.json")

# Load image and JSON
img = cv2.imread(img_path)

with open(json_path, 'r') as f:
    data = json.load(f)

# Draw bounding boxes
for piece in data['pieces']:
    x, y, w, h = piece['box']
    label = piece['piece']
    # Draw rectangle
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    # Put label
    cv2.putText(img, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

# Save and show
# cv2.imwrite('0000_bb.png', img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
